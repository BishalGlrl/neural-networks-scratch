```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class RNN:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.01):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.lr = learning_rate
        
        # Initialize weights
        self.Wxh = np.random.randn(hidden_size, input_size) * 0.01  # Input to hidden
        self.Whh = np.random.randn(hidden_size, hidden_size) * 0.01  # Hidden to hidden
        self.Why = np.random.randn(output_size, hidden_size) * 0.01  # Hidden to output
        
        # Initialize biases
        self.bh = np.zeros((hidden_size, 1))
        self.by = np.zeros((output_size, 1))
    
    def tanh(self, x):
        return np.tanh(x)
    
    def tanh_derivative(self, x):
        return 1 - x ** 2
    
    def softmax(self, x):
        exp_x = np.exp(x - np.max(x))
        return exp_x / exp_x.sum(axis=0, keepdims=True)
    
    def forward(self, inputs):
        h = np.zeros((self.hidden_size, 1))
        self.inputs = inputs
        self.hidden_states = [h]
        self.outputs = []
        
        # Forward through time
        for x in inputs:
            # Reshape input
            x = x.reshape(-1, 1)
            
            # Update hidden state
            h = self.tanh(np.dot(self.Wxh, x) + np.dot(self.Whh, h) + self.bh)
            self.hidden_states.append(h)
            
            # Compute output
            y = np.dot(self.Why, h) + self.by
            self.outputs.append(y)
        
        return self.outputs
    
    def backward(self, targets):
        # Initialize gradients
        dWxh, dWhh, dWhy = np.zeros_like(self.Wxh), np.zeros_like(self.Whh), np.zeros_like(self.Why)
        dbh, dby = np.zeros_like(self.bh), np.zeros_like(self.by)
        
        # Gradient through time
        dh_next = np.zeros((self.hidden_size, 1))
        
        # Backward through time
        for t in reversed(range(len(self.inputs))):
            # Output gradient
            dy = self.outputs[t].copy()
            target = targets[t].reshape(-1, 1)
            dy -= target  # Gradient of softmax cross-entropy
            
            # Hidden to output gradients
            dWhy += np.dot(dy, self.hidden_states[t+1].T)
            dby += dy
            
            # Hidden state gradient
            dh = np.dot(self.Why.T, dy) + dh_next
            
            # Gradient through tanh
            dh_raw = self.tanh_derivative(self.hidden_states[t+1]) * dh
            
            # Input to hidden gradients
            dWxh += np.dot(dh_raw, self.inputs[t].reshape(1, -1))
            
            # Hidden to hidden gradients
            dWhh += np.dot(dh_raw, self.hidden_states[t].T)
            dbh += dh_raw
            
            # Pass gradient to previous time step
            dh_next = np.dot(self.Whh.T, dh_raw)
        
        # Clip gradients to prevent exploding
        for grad in [dWxh, dWhh, dWhy, dbh, dby]:
            np.clip(grad, -5, 5, out=grad)
        
        return dWxh, dWhh, dWhy, dbh, dby
    
    def update(self, grads):
        dWxh, dWhh, dWhy, dbh, dby = grads
        
        # Update parameters with gradient descent
        self.Wxh -= self.lr * dWxh
        self.Whh -= self.lr * dWhh
        self.Why -= self.lr * dWhy
        self.bh -= self.lr * dbh
        self.by -= self.lr * dby
    
    def train(self, training_data, epochs=100):
        losses = []
        
        for epoch in range(epochs):
            total_loss = 0
            
            for inputs, targets in training_data:
                # Forward pass
                outputs = self.forward(inputs)
                
                # Calculate loss
                loss = 0
                for t in range(len(outputs)):
                    target = targets[t].reshape(-1, 1)
                    output = outputs[t]
                    loss += -np.sum(target * np.log(self.softmax(output) + 1e-8))
                
                total_loss += loss
                
                # Backward pass
                grads = self.backward(targets)
                
                # Update parameters
                self.update(grads)
            
            avg_loss = total_loss / len(training_data)
            losses.append(avg_loss)
            
            if epoch % 10 == 0:
                print(f"Epoch {epoch}, Loss: {avg_loss:.4f}")
        
        return losses
    
    def predict(self, inputs):
        outputs = self.forward(inputs)
        predictions = []
        
        for output in outputs:
            probs = self.softmax(output)
            predictions.append(np.argmax(probs))
        
        return predictions
```

```python
def create_sine_wave_data(seq_length=50, num_samples=100):
    """Create sine wave prediction dataset"""
    data = []
    
    for _ in range(num_samples):
        # Generate sine wave
        freq = np.random.uniform(0.1, 0.5)
        phase = np.random.uniform(0, 2*np.pi)
        
        # Input: current value
        # Target: next value
        time_steps = np.arange(seq_length + 1)
        sine_wave = np.sin(freq * time_steps + phase)
        
        inputs = sine_wave[:-1].reshape(-1, 1)
        targets = sine_wave[1:].reshape(-1, 1)
        
        # Convert to classification (4 classes based on quadrant)
        input_classes = np.digitize(inputs, [-0.5, 0, 0.5])
        target_classes = np.digitize(targets, [-0.5, 0, 0.5])
        
        # One-hot encode
        input_onehot = np.eye(4)[input_classes.flatten()]
        target_onehot = np.eye(4)[target_classes.flatten()]
        
        data.append((input_onehot, target_onehot))
    
    return data

if __name__ == "__main__":
    # Create toy dataset
    data = create_sine_wave_data(seq_length=20, num_samples=50)
    
    # Create RNN
    rnn = RNN(input_size=4, hidden_size=16, output_size=4, learning_rate=0.01)
    
    # Train RNN
    losses = rnn.train(data, epochs=100)
    
    # Plot results
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(losses)
    plt.title('Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    
    # Test prediction
    plt.subplot(1, 2, 2)
    test_inputs, test_targets = data[0]
    predictions = rnn.predict(test_inputs)
    
    # Convert one-hot back to values
    actual = np.argmax(test_targets, axis=1)
    
    plt.plot(actual, label='Actual', marker='o')
    plt.plot(predictions, label='Predicted', marker='x', linestyle='--')
    plt.title('Sequence Prediction')
    plt.xlabel('Time Step')
    plt.ylabel('Class')
    plt.legend()
    plt.tight_layout()
    
    plt.savefig('rnn_results.png')
    plt.show()
```
